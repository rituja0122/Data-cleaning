##### To fill categorial values

```python
for i in data.select_dtypes(include="object").columns:
    data[i].fillna(data[i].mode()[0],inplace=True)
```

```python
data.isnull().sum() 
#only numerical values have remained
```

##### Ordinal Encoding

```python
dframe = pd.DataFrame({"Size":["s","m","l","xl","s","m","l","xl","s","xl","m"]})
dframe
```

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
ord_data = [["s","m","l","xl"]]
```

```python
oe = OrdinalEncoder(categories=ord_data)
oe.fit(dframe[["Size"]])
```

```python
dframe["en_size"] = oe.transform(dframe[["Size"]])
dframe
```

##### Ordinal Encoding through map

```python
ord_data1 = {"s":0,"m":1,"l":2,"xl":3}
```

```python
dframe["en_size1"] = dframe["Size"].map(ord_data1)
dframe
```
