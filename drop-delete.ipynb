##### To drop null values

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("food_coded.csv")
```

##### To delete a particular column/columns

```python
data.drop(columns=["type_sports"],inplace=True)
#inplace modifies in original dataset
data.head(3)
```

```python
data.shape
```

##### To delete all rows containing null values

```python
data.dropna(inplace=True)
```

```python
data.shape
```

```python
sns.heatmap(data.isnull())
plt.show()
```
