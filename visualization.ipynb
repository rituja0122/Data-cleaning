```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

##### Feature Scaling

```python
# Feature Scaling is of two types
# 1. Standardization
# 2. Normalization
```

##### Standardization

```python
# dataset(x)
# mean(u)
# standard deviation(d)

# x_new = (x - u)/d

# after standardization mean becomes 0 and variance become 1
```

```python
data = pd.read_csv("food_coded.csv")
data.head(3)
```

```python
data["calories_scone"].isnull().sum()
```

```python
data["calories_scone"].fillna(data["calories_scone"].mean(),inplace=True)
```

```python
data["calories_scone"].isnull().sum()
```

```python
sns.distplot(data["calories_scone"])
plt.show()
```

```python
data.describe()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
ss = StandardScaler()
ss.fit(data[["calories_scone"]])
```

```python
data["calories_scone_ss"] = pd.DataFrame(ss.transform(data[["calories_scone"]]))
data.head(3)
```

```python
data.describe()
```

```python
plt.subplot(1,2,1)
plt.title("Before")
sns.distplot(data["calories_scone"])

plt.subplot(1,2,2)
plt.title("After")
sns.distplot(data["calories_scone_ss"])

plt.show()
```

##### Normalization

###### Min-Max scaling

```python
# In this scaling technique the values are rescaled within the range of 0 to 1

# x_new = (x - min(x))/(max(x) - min(x))
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
ms = MinMaxScaler()
ms.fit(data[["calories_scone"]])
```

```python
data["calories_scone_ms"] = pd.DataFrame(ms.transform(data[["calories_scone"]]))
data.head(3)
```

```python
plt.figure(figsize=(10,7))

plt.subplot(1,2,1)
plt.title("Before")
sns.distplot(data["calories_scone"])

plt.subplot(1,2,2)
plt.title("After")
sns.distplot(data["calories_scone_ms"])

plt.show()
```

```python
plt.subplot(1,3,1)
plt.title("Original")
sns.distplot(data["calories_scone"])

plt.subplot(1,3,2)
plt.title("Standardization")
sns.distplot(data["calories_scone_ss"])

plt.subplot(1,3,3)
plt.title("Normalization")
sns.distplot(data["calories_scone_ms"])

plt.show()
```
